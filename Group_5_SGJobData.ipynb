```python
file_path = './SGJobData.csv'
df = pd.read_csv(file_path)

display(df.head())
```

```python
print("Number of rows:", df.shape[0])
```

```python
# Convert date columns to datetime objects
date_columns = ['metadata_expiryDate', 'metadata_newPostingDate', 'metadata_originalPostingDate']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Create a new column for the average salary
df['average_salary'] = (df['salary_minimum'] + df['salary_maximum']) / 2

display(df[['salary_minimum', 'salary_maximum', 'average_salary', 'metadata_newPostingDate']].head())
```

```python
# Display column names and data types
print("Column Information:")
display(df.info())

# Display basic statistics for numerical columns
print("\nBasic Statistics:")
display(df.describe())

# Display the number of unique values in each column
print("\nUnique Values per Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
import plotly.express as px

# Count job postings by date
job_postings_over_time = df.groupby(df['metadata_newPostingDate'].dt.date).size().reset_index(name='count')
job_postings_over_time['metadata_newPostingDate'] = pd.to_datetime(job_postings_over_time['metadata_newPostingDate'])

# Create an interactive time series plot
fig = px.line(job_postings_over_time, x='metadata_newPostingDate', y='count', title='Job Postings Over Time')
fig.show()
```

```python
import plotly.graph_objects as go

# Group by date and calculate average salary
salary_over_time = df.groupby(df['metadata_newPostingDate'].dt.date)[['salary_minimum', 'salary_maximum']].mean().reset_index()
salary_over_time['metadata_newPostingDate'] = pd.to_datetime(salary_over_time['metadata_newPostingDate'])

# Create an interactive time series plot for average salaries
fig = go.Figure()

fig.add_trace(go.Scatter(x=salary_over_time['metadata_newPostingDate'], y=salary_over_time['salary_minimum'], mode='lines', name='Average Minimum Salary'))
fig.add_trace(go.Scatter(x=salary_over_time['metadata_newPostingDate'], y=salary_over_time['salary_maximum'], mode='lines', name='Average Maximum Salary'))


fig.update_layout(
    title='Average Minimum and Maximum Salary Over Time',
    xaxis_title='Date',
    yaxis_title='Salary'
)

fig.show()
```

```python
# Analyze the 'categories' column to identify tech-related entries
# Due to the format of the 'categories' column (list of dictionaries as a string),
# we need to process it to extract the category names.
import ast

def extract_categories(categories_str):
    try:
        categories_list = ast.literal_eval(categories_str)
        return [item['category'] for item in categories_list]
    except (ValueError, SyntaxError, TypeError):
        return []

df['category_list'] = df['categories'].apply(extract_categories)

# Get all unique categories
all_categories = [item for sublist in df['category_list'].dropna() for item in sublist]
unique_categories = pd.Series(all_categories).unique()

print("Unique Categories:")
display(unique_categories)

# Function to identify tech jobs based on keywords in the job title
def is_tech_job_by_title(title):
    if isinstance(title, str):
        tech_keywords = ['tech', 'software', 'developer', 'engineer', 'data scientist', 'analyst', 'IT', 'information technology', 'cybersecurity', 'network', 'cloud', 'AI', 'machine learning', 'developer', 'programmer']
        for keyword in tech_keywords:
            if keyword in title.lower():
                return True
    return False

# Create a new column 'is_tech_title' based on the job title
df['is_tech_title'] = df['title'].apply(is_tech_job_by_title)

# Display the number of tech vs non-tech jobs based on title
print("\nNumber of tech-related jobs (based on title):", df['is_tech_title'].sum())
print("Number of non-tech jobs (based on title):", len(df) - df['is_tech_title'].sum())

# Display the first few rows with the new 'is_tech_title' column
display(df[['title', 'is_tech_title']].head())
```

```python
# Filter for tech-related jobs based on the 'is_tech_title' column
tech_jobs_df = df[df['is_tech_title'] == True]

# Group by company name and count the number of tech job postings
company_tech_job_counts = tech_jobs_df['postedCompany_name'].value_counts().reset_index()
company_tech_job_counts.columns = ['Company', 'Tech Job Count']

# Display the top 10 companies hiring the most tech jobs
print("Top 10 Companies Hiring the Most Tech Jobs:")
display(company_tech_job_counts.head(10))
```

## Extract and normalize categories

### Subtask:
Process the 'categories' column to extract and normalize the industry sectors.


**Reasoning**:
Define a function to extract categories, apply it to the 'categories' column, explode the resulting list column, and normalize the category names.



```python
import ast

def extract_categories_safe(categories_str):
    """Safely extracts category names from the categories string."""
    try:
        categories_list = ast.literal_eval(categories_str)
        return [item['category'] for item in categories_list if isinstance(item, dict) and 'category' in item]
    except (ValueError, SyntaxError, TypeError):
        return []

df['category_list_extracted'] = df['categories'].apply(extract_categories_safe)

# Explode the list of categories
df_exploded = df.explode('category_list_extracted')

# Clean and normalize category names
df_exploded['category_list_extracted'] = df_exploded['category_list_extracted'].str.strip().str.lower()

# Display the first few rows of the exploded DataFrame with the new normalized category column
display(df_exploded[['categories', 'category_list_extracted']].head())
```

## Calculate total postings per sector

### Subtask:
Group the data by sector and count the total number of job postings for each sector.


**Reasoning**:
Group the data by sector and count the total number of job postings for each sector, then rename the columns and display the result.



```python
# Group by sector and count total postings
sector_postings = df_exploded.groupby('category_list_extracted').size().reset_index(name='Total Postings')

# Rename columns
sector_postings.columns = ['Sector', 'Total Postings']

# Display the first few rows
display(sector_postings.head())
```

## Analyze growth trend per sector

### Subtask:
Analyze the trend of new vs. older postings within each sector to identify growth or decline.


**Reasoning**:
Calculate posting age, group by sector, calculate average posting age, rename columns, and display the result.



```python
from datetime import datetime

# Calculate posting recency in days
today = datetime.now()
df_exploded['posting_age'] = (today - df_exploded['metadata_newPostingDate']).dt.days

# Group by sector and calculate average posting age
sector_growth_trend = df_exploded.groupby('category_list_extracted')['posting_age'].mean().reset_index()

# Rename columns
sector_growth_trend.columns = ['Sector', 'Average Posting Age']

# Display the first few rows of the resulting DataFrame
display(sector_growth_trend.head())
```

## Calculate median salary per sector

### Subtask:
Determine the median salary for each industry sector.


**Reasoning**:
Group the data by sector and calculate the median salary, then rename the columns and display the result.



```python
# Group by sector and calculate median salary
sector_median_salary = df_exploded.groupby('category_list_extracted')['average_salary'].median().reset_index()

# Rename columns
sector_median_salary.columns = ['Sector', 'Median Salary']

# Display the first few rows
display(sector_median_salary.head())
```

## Identify dominant roles per sector

**Reasoning**:
Group the data by sector and find the value counts of job titles within each sector, then select the top 10 most frequent job titles for each sector and display the results for a few example sectors.



```python
# Group by sector and find the value counts of job titles within each sector
dominant_roles_per_sector = df_exploded.groupby('category_list_extracted')['title'].apply(lambda x: x.value_counts().head(10)).reset_index()

# Rename columns
dominant_roles_per_sector.columns = ['Sector', 'Job Title', 'Count']

# Display the dominant roles for a few example sectors
example_sectors = ['information technology', 'healthcare / pharmaceutical', 'education and training']
for sector in example_sectors:
    print(f"\nDominant Roles in '{sector}':")
    display(dominant_roles_per_sector[dominant_roles_per_sector['Sector'] == sector])
```
